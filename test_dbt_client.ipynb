```python
from pyveb import dbt_client

dbt = dbt_client.dbtClient(
    "https://dh138.us1.dbt.com/api/", 
    "v2",
    38487,
    env = 'prd',
    auth_type = 'api_key'
)
```

```python
conns = dbt.get_connections()
print(f'found {len(conns)} number of connections')


```
